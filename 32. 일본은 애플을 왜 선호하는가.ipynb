```r
JapanApple <- read.csv("./Data/JapanApple.csv")
head(JapanApple)

# 애플의 제품이 전체 판매대수에 영향을 끼치는가? 
# - 총판매대수의 75%의 영향을 가지고 있다.
cor(JapanApple$All, JapanApple$Apple) 

All <- JapanApple$All
Apple <- JapanApple$Apple

Apple2 <- cbind(All, Apple)
Apple2

chisq.test(Apple2)
```

```r
# 파이 그래프 임포트 
library(plotrix)

# 2018년도 전체 판매대수
JapanApple2018 <- JapanApple[JapanApple$년도== "2018",]
JapanApple2018 

# 2019년도 전체 판매대수
JapanApple2019 <- JapanApple[JapanApple$년도== "2019",]
JapanApple2019 

# 2020년도 전체 판매대수
JapanApple2020 <- JapanApple[JapanApple$년도== "2020",]
JapanApple2020 

# 2021년도 전체 판매대수
JapanApple2021 <- JapanApple[JapanApple$년도== "2021",]
JapanApple2021 

# 2022년도 전체 판매대수
JapanApple2022 <- JapanApple[JapanApple$년도== "2022",]
JapanApple2022 


# 2018년도 Ratio
JapanApple2018_ratio <- round(c(JapanApple2018$Apple, JapanApple2018$Sharp, JapanApple2018$Sony, JapanApple2018$Samsung, 
JapanApple2018$Kyocera, JapanApple2018$Fujitsu) / JapanApple2018$All * 100,2)
JapanApple2018_ratio

# 2019년도 Ratio
JapanApple2019_ratio <- round(c(JapanApple2019$Apple, JapanApple2019$Sharp, JapanApple2019$Sony, JapanApple2019$Samsung, 
JapanApple2019$Kyocera, JapanApple2019$Fujitsu) / JapanApple2019$All * 100,2)
JapanApple2019_ratio

# 2020년도 Ratio
JapanApple2020_ratio <- round(c(JapanApple2020$Apple, JapanApple2020$Sharp, JapanApple2020$Sony, JapanApple2020$Samsung, 
JapanApple2020$Kyocera, JapanApple2020$Fujitsu) / JapanApple2020$All * 100,2)
JapanApple2020_ratio

# 2021년도 Ratio
JapanApple2021_ratio <- round(c(JapanApple2021$Apple, JapanApple2021$Sharp, JapanApple2021$Sony, JapanApple2021$Samsung, 
JapanApple2021$Kyocera, JapanApple2021$Fujitsu) / JapanApple2021$All * 100,2)
JapanApple2021_ratio

# 2022년도 Ratio
JapanApple2022_ratio <- round(c(JapanApple2022$Apple, JapanApple2022$Sharp, JapanApple2022$Sony, JapanApple2022$Samsung, 
JapanApple2022$Kyocera, JapanApple2022$Fujitsu) / JapanApple2022$All * 100,2)
JapanApple2022_ratio

# 2018 Label 만들기  (제목 데이터순서대로 이름을 지어준다.)
JapanApple2018_label <-
paste(c("Apple", "Sharp", "Sony", "Samsung", "Kyocera", "Fujitsu" ),"\n", JapanApple2018_ratio, "%") 
JapanApple2018_label

# 2019 Label 만들기  (제목 데이터순서대로 이름을 지어준다.)
JapanApple2019_label <-
paste(c("Apple", "Sharp", "Sony", "Samsung", "Kyocera", "Fujitsu" ),"\n", JapanApple2019_ratio, "%") 
JapanApple2019_label

# 2020 Label 만들기  (제목 데이터순서대로 이름을 지어준다.)
JapanApple2020_label <-
paste(c("Apple", "Sharp", "Sony", "Samsung", "Kyocera", "Fujitsu" ),"\n", JapanApple2020_ratio, "%") 
JapanApple2020_label

# 2021 Label 만들기  (제목 데이터순서대로 이름을 지어준다.)
JapanApple2021_label <-
paste(c("Apple", "Sharp", "Sony", "Samsung", "Kyocera", "Fujitsu" ),"\n", JapanApple2021_ratio, "%") 
JapanApple2021_label

# 2022 Label 만들기  (제목 데이터순서대로 이름을 지어준다.)
JapanApple2022_label <-
paste(c("Apple", "Sharp", "Sony", "Samsung", "Kyocera", "Fujitsu" ),"\n", JapanApple2022_ratio, "%") 
JapanApple2022_label


# 2019 판매량별 정렬 
sorted_data <- sort(JapanApple2019_ratio, decreasing = TRUE)
sorted_labels <- JapanApple2019_label[order(JapanApple2019_ratio, decreasing = TRUE)]

# 2020 판매량별 정렬 
sorted_data2020 <- sort(JapanApple2020_ratio, decreasing = TRUE)
sorted_labels2020 <- JapanApple2020_label[order(JapanApple2020_ratio, decreasing = TRUE)]

# 2021 판매량별 정렬 
sorted_data2021 <- sort(JapanApple2021_ratio, decreasing = TRUE)
sorted_labels2021 <- JapanApple2021_label[order(JapanApple2021_ratio, decreasing = TRUE)]


# pie 그래프  
par(mfrow = c(2,2), bg="white", family="AppleGothic")

pie(JapanApple2018_ratio, 
    main = "2018년도 일본의 스마트폰 판매율",
    labels = JapanApple2018_label,
    col = c("DarkGray", "Thistle", "Ivory", "LightSlateGray", "skyblue", "MistyRose"),
    border = 40
)
pie(sorted_data, 
    main = "2019년도 일본의 스마트폰 판매율",
    labels = sorted_labels,
    col = c("DarkGray", "Thistle", "Ivory", "LightSlateGray", "skyblue", "MistyRose"),
    border = 40
)

pie(sorted_data2020, 
    main = "2020년도 일본의 스마트폰 판매율",
    labels = sorted_labels2020,
    col = c("DarkGray", "Thistle", "Ivory", "LightSlateGray", "skyblue", "MistyRose"),
    border = 40
)

pie(sorted_data2021, 
    main = "2021년도 일본의 스마트폰 판매율",
    labels = sorted_labels2021,
    col = c("DarkGray", "Thistle", "Ivory", "LightSlateGray", "skyblue", "MistyRose"),
    border = 40
)
```

```r
# 판매량 막대그래프 

library(ggplot2)

# 데이터프레임 생성
JapanApple2022_gg <- data.frame(label = JapanApple2022_label, ratio = JapanApple2022_ratio)

# 판매량 기준으로 데이터프레임 정렬
JapanApple2022_gg <- JapanApple2022_gg[order(JapanApple2022_gg$ratio, decreasing = TRUE), ]

par(bg="white", family="AppleGothic")

# 가로 막대 그래프 그리기
ggplot(JapanApple2022_gg, aes(x = reorder(label, ratio), y = ratio, fill = label)) +
  geom_bar(stat = "identity", width = 0.5, color = "black") +
  geom_text(aes(label = paste0(ratio, "%")), vjust = 5, size = 5, color = "black") +
  theme_minimal() +
  labs(title = "2022년도 일본의 스마트폰 판매율") +
  scale_fill_manual(values = c("DarkGray", "Thistle", "Ivory", "LightSlateGray", "skyblue", "MistyRose")) +
  xlab(NULL) +
  ylab(NULL) +
  coord_flip() +
  theme(plot.title = element_text(family = "AppleGothic"))
        


 
```
